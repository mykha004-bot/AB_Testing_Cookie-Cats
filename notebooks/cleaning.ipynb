```python
cookie_cats_cleaned = cookie_cats_cleaned.dropna()
```

```python
cookie_cats_cleaned.info()
```

```python
cookie_cats_cleaned['sum_gamerounds'].sort_values(ascending=False).head(10)
```

```python
Q1 = cookie_cats_cleaned['sum_gamerounds'].quantile(0.25)
Q3 = cookie_cats_cleaned['sum_gamerounds'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = cookie_cats_cleaned[(cookie_cats_cleaned['sum_gamerounds'] < lower_bound) | (cookie_cats_cleaned['sum_gamerounds'] > upper_bound)]
```

```python
cookie_cats_cleaned = cookie_cats_cleaned[cookie_cats_cleaned['sum_gamerounds'] <= 45000].copy()

```

```python
cookie_cats_cleaned.to_csv("../data/cookie_cats_cleaned.csv", index=False)
```
